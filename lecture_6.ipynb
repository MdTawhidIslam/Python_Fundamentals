**Problem 1: Power of Two**

Given an integer n, return true if it is a power of two. Otherwise, return false.

An integer n is a power of two, if there exists an integer x such that n == 2^x.

**Example 1:**
<br>Input: n = 1
<br>Output: true
<br>Explanation: 2^0 = 1

**Example 2:**
<br>Input: n = 16
<br>Output: true
<br>Explanation: 2^4 = 16

**Example 3:**
<br>Input: n = 3
<br>Output: false

**Example 4:**
<br>Input: n = 4
<br>Output: true

**Example 5:**
<br>Input: n = 5
<br>Output: false

n == 2^x
<br>=> log(n) = log(2^x)
<br>=> log(n) = x*log(2)
<br>=> x = log(n)/log(2)

```python
import math

def isPowerOfTwo(n):

    if n <= 0:
        return False

    x = math.log10(n)/math.log10(2)
    return n == 2**int(x)
```

```python
p1_example_1 = isPowerOfTwo(1)
print(p1_example_1)

print("-----------------------------")

p1_example_2 = isPowerOfTwo(16)
print(p1_example_2)

print("-----------------------------")

p1_example_3 = isPowerOfTwo(3)
print(p1_example_3)

print("-----------------------------")

p1_example_4 = isPowerOfTwo(4)
print(p1_example_4)

print("-----------------------------")

p1_example_5 = isPowerOfTwo(5)
print(p1_example_5)
```

**Problem 3: Two Sum**

Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.

You may assume that each input would have exactly one solution, and you may not use the same element twice.

You can return the answer in any order.

**Example 1:**
<br>Input: nums = [2,7,11,15], target = 9
<br>Output: [0,1]
<br>Output: Because nums[0] + nums[1] == 9, we return [0, 1].

**Example 2:**
<br>Input: nums = [3,2,4], target = 6
<br>Output: [1,2]

**Example 3:**
<br>Input: nums = [3,3], target = 6
<br>Output: [0,1]

```python
def twoSum(nums, target):

    val_index_dict = {}

    for index in range(0, len(nums)):
        goal = target - nums[index]
        print("x = ", goal, "y = ", nums[index])

        if(goal in val_index_dict):
            print([val_index_dict[goal], index])
            return [val_index_dict[goal], index]

        val_index_dict[nums[index]] = index
        print("dictionary = ", val_index_dict)
```

```python
p3_example_1 = twoSum([2,7,11,15], 9)
print(p3_example_1)

print("-----------------------------")

p3_example_2 = twoSum([3,2,4], 6)
print(p3_example_2)

print("-----------------------------")

p3_example_3 = twoSum([3,3], 6)
print(p3_example_3)
```

**Problem 4: Roman to Integer**

Roman numerals are represented by seven different symbols: I, V, X, L, C, D and M.

values = {"I": 1,"V": 5,"X": 10,"L": 50,"C": 100,"D": 500,"M": 1000}

For example, 2 is written as II in Roman numeral, just two one's added together. 12 is written as XII, which is simply X + II. The number 27 is written as XXVII, which is XX + V + II.

Roman numerals are usually written largest to smallest from left to right. However, the numeral for four is not IIII. Instead, the number four is written as IV. Because the one is before the five we subtract it making four. The same principle applies to the number nine, which is written as IX. There are six instances where subtraction is used:

I can be placed before V (5) and X (10) to make 4 and 9. 
<br>X can be placed before L (50) and C (100) to make 40 and 90. 
<br>C can be placed before D (500) and M (1000) to make 400 and 900.
<br>Given a roman numeral, convert it to an integer.

**Example 1:**
<br>Input: s = "III"
<br>Output: 3

**Example 2:**
<br>Input: s = "IV"
<br>Output: 4

**Example 3:**
<br>Input: s = "IX"
<br>Output: 9

**Example 4:**
<br>Input: s = "LVIII"
<br>Output: 58
<br>Explanation: L = 50, V= 5, III = 3.

**Example 5:**
<br>Input: s = "MCMXCIV"
<br>Output: 1994
<br>Explanation: M = 1000, CM = 900, XC = 90 and IV = 4.
 
**Constraints:**
1. 1 <= s.length <= 15
2. s contains only the characters ('I', 'V', 'X', 'L', 'C', 'D', 'M').
3. It is guaranteed that s is a valid roman numeral in the range [1, 3999].

```python
values = {"I": 1,"V": 5,"X": 10,"L": 50,"C": 100,"D": 500,"M": 1000}

def romanToInt(s):

    total = 0
    i = 0 #index
    
    while i < len(s):
        
        # If this is the subtractive case.
        if i + 1 < len(s) and values[s[i]] < values[s[i + 1]]:

            print("\nsubtractive case")
            print("current digit, ", s[i], " = ", values[s[i]])
            print("next digit, ", s[i+1], " = ", values[s[i+1]])

            total += values[s[i + 1]] - values[s[i]]

            print('total_by_subtraction = ', total)
            print("----------------------")
            i += 2
            
        # Else this is NOT the subtractive case.
        else:
            print("current digit, ", s[i], " = ", values[s[i]])
            total += values[s[i]]
            print('total_by_addition = ', total)
            print("----------------------")
            i += 1
            
        print()
    return total 
```

```python
p4_example_1 = romanToInt("III")
print(p4_example_1)

print("-----------------------------------------------")

p4_example_2 = romanToInt("IV")
print(p4_example_2)

print("-----------------------------------------------")

p4_example_3 = romanToInt("IX")
print(p4_example_3)

print("-----------------------------------------------")

p4_example_4 = romanToInt("LVIII")
print(p4_example_4)

print("-----------------------------------------------")

p4_example_5 = romanToInt("MCMXCIV")
print(p4_example_5)
```

**Problem 5: Unique Morse Code Words**

International Morse Code defines a standard encoding where each letter is mapped to a series of dots and dashes, as follows: "a" maps to ".-", "b" maps to "-...", "c" maps to "-.-.", and so on.

For convenience, the full table for the 26 letters of the English alphabet is given below:

[".-","-...","-.-.","-..",".","..-.","--.","....","..",".---","-.-",".-..","--","-.","---",".--.","--.-",".-.","...","-","..-","...-",".--","-..-","-.--","--.."]

Now, given a list of words, each word can be written as a concatenation of the Morse code of each letter. For example, "cab" can be written as "-.-..--...", (which is the concatenation "-.-." + ".-" + "-..."). We'll call such a concatenation, the transformation of a word.

Return the number of different transformations among all words we have.

**Example:**
<br>Input: words = ["gin", "zen", "gig", "msg"]
<br>Output: 2
<br>Explanation: 
<br>The transformation of each word is:
<br>"gin" -> "--...-."
<br>"zen" -> "--...-."
<br>"gig" -> "--...--."
<br>"msg" -> "--...--."

There are 2 different transformations, "--...-." and "--...--.".


```python
def uniqueMorseRepresentations(words):

    morse = {'a':".-", 'b':"-...", 'c':"-.-.", 'd':"-..", 'e':".",
            'f':"..-.", 'g':"--.", 'h':"....", 'i':"..", 'j':".---",
            'k':"-.-", 'l':".-..", 'm':"--", 'n':"-.", 'o':"---",
             'p':".--.", 'q':"--.-", 'r':".-.", 's':"...", 't':"-",
             'u':"..-", 'v':"...-", 'w':".--", 'x':"-..-", 'y':"-.--",
             'z':"--.."}

    morse_signal = ""
    morse_list = []
    
    for word in words:
        for char in word:
            morse_signal += morse[char]

        morse_list.append(morse_signal)
        morse_signal = ""
        print(morse_list)

    return len(set(morse_list))
```

```python
p5_example = uniqueMorseRepresentations(["gin", "zen", "gig", "msg"])
print(p5_example)
```
